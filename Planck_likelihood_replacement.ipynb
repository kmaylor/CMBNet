```python
def planckdic(c):
    c['ommh2']=c['omegabh2']+c['omegach2']
    c['As'] = exp(c['logA'])*10**(-10)
    c['100ombh2']=100*c['omegabh2']
    c['theta_s']=c['thetastar*']
    return c

def slik2cmc( c ):
    c['ns'] = c['cosmo.ns']
    c['omegabh2'] = c['cosmo.ombh2']
    c['omegach2'] = c['cosmo.omch2']
    c['tau'] = c['cosmo.tau']
    c['theta_s'] = c['cosmo.theta']*100   
    try:
        c['clamp*'] = exp(c['cosmo.logA'])*exp(-2*c['tau'])/10
    except KeyError:
        c['clamp*'] = c['cosmo.As']*exp(-2*c['tau'])*1e9
    c['ommh2']=c['omegabh2']+c['omegach2']
    c['H0*'] = c['cosmo.H0']
    #c['Yp'] = c['cosmo.Yp']
    c['100ombh2']=100*c['omegabh2']
    try:
        c['cal']=c['sptsz.cal']
    except KeyError:
        c['cal'] = c['s12.cal']
    return c
```

```python
Pchain_file="/home/kmaylor/Python_Projects/PlanckVSPT/Planck_2015_parameters/base/plikHM_TT_lowTEB/base_plikHM_TT_lowTEB" #loading planck 2015 chain
Pchain=chains.load_cosmomc_chain(Pchain_file).burnin(1000).join()
Pchain=planckdic(Pchain)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Activation
```

```python
model = Sequential([
    Dense(12, input_shape=(6,)),
    Activation('sigmoid'),
    Dense(1),
    Activation('relu')
])
```

```python
model.compile(optimizer='rmsprop',
              loss='mse')
```

```python
c=round(len(Pchain['H0*'])/2)
training_data=array([ i for i in zip(*[Pchain['H0*'][:c],
                              Pchain['ommh2'][:c]*10,Pchain['omegabh2'][:c]*100,
                              Pchain['tau'][:c]*100,Pchain['clamp*'][:c],Pchain['ns'][:c]*10])])
training_labels=array(Pchain['lnl'][:c])
```

```python
model.fit(training_data, training_labels, epochs=1000, batch_size=10)
```

```python
test_data=array([ i for i in zip(*[Pchain['H0*'][c:],
                              Pchain['ommh2'][c:]*10,Pchain['omegabh2'][c:]*100,
                              Pchain['tau'][c:]*100,Pchain['clamp*'][c:],Pchain['ns'][c:]*10])])
test_labels=array(Pchain['lnl'][c:])
```

```python
sum(abs([m[0] for m in model.predict(test_data)]-test_labels)/test_labels)/len(test_labels)
```
